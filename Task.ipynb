# TASK : Students Mark Analysis

# 1. Find the average by students and subject

```python
import numpy as np
```

```python
np.random.seed(83)
marks = np.random.randint(20,100,(10,5))
marks
```

axis = 0- subject(column), axis = 1 - students(row)

```python
# average by students(row-wise)
student_avg = np.mean(marks,axis =1)

# average by subjects((column-wise)
subject_avg = np.mean(marks,axis=0)

print("Student Average:",student_avg)

print("\nSubject Average:",subject_avg)
```

# 2. Find highest and lowest per subject

```python
# Highest mark per subject
highest_mark = np.max(marks,axis=0)

# lowest mark per subject
lowest_mark = np.min(marks,axis=0)

print("Highest mark per subject:",highest_mark)
print("\nLowest mark per subject",lowest_mark)
```

# 3.Find which student is overall class topper

```python
# sum of all students marks
std_total = np.sum(marks,axis = 1)

# Overall class topper
class_topper = np.max(std_total)
class_topper
```

# 4.Find the total students have passed by counting the subject

```python
# count students passed in each subject
pass_count = np.sum(marks >= 36, axis = 0)
pass_count

# display result
for i,count in enumerate(pass_count,start=1):
    print(f"Subject{i} {count} students passed")
```

# 5. Find which subject is difficult

```python
# count students fail in particular subject
fail_count= np.sum(marks<= 35,axis = 0)
fail_count

# display result
for i,count in enumerate (fail_count,start = 1):
    print(f"Subjects {i} {count} students failed")

# find difficult subject 
difficult_sub = np.argmax(fail_count)
print(f"Most difficult subject is: Subject {difficult_sub + 1}")
```
