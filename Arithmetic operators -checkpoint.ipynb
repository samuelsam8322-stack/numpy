# ARITHMETIC OPERATORS AND FUNCTIONS:

# 1. arange() → Create Arrays

np.arange() creates arrays with a sequence of numbers.

```python
import numpy as np
```

```python
np.arange(start, stop, step)
```

```python
start → starting value
stop → ending value (not included)
step → gap between numbers
```

```python
import numpy as np

a = np.arange(1, 10, 2)
print(a)
```

# 2. add() → Addition

Adds two arrays element by element.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.add(a, b))
```

```python
a + b
```

# 3. subtract() → Subtraction

Subtracts elements.

```python
print(np.subtract(a, b))
```

```python
a - b
```

# 4. multiply() → Multiplication

Multiplies element by element.

```python
print(np.multiply(a, b))
```

```python
a * b
```

# 5. divide() → Division

Divides elements.

```python
print(np.divide(b, a))
```

```python
b / a
```

# 6. power() → Exponent / Square / Cube

Raises numbers to powers.

```python
a = np.array([2, 3, 4])

print(np.power(a, 2))
```

```python
a ** 2
```

# 7. mod() → Remainder

Returns remainder after division.

```python
a = np.array([10, 20, 30])

print(np.mod(a, 3))
```

```python
a % 3
```

# 8. round() → Round Decimal Values

Rounds numbers to nearest value.

```python
a = np.array([2.3, 4.7, 5.5])

print(np.round(a))
```

```python
# Round to 2 decimal places
a = np.array([1.2345, 2.5678])

print(np.round(a, 2))
```

# 9. ceil() → Round Up

Always rounds UP to nearest integer.

```python
a = np.array([1.2, 2.1, 3.9])

print(np.ceil(a))
```

# 10. floor() → Round Down

Always rounds DOWN.

```python
print(np.floor(a))
```

# One Combined Example

```python
import numpy as np

a = np.arange(1, 6)
b = np.array([10, 20, 30, 40, 50])

print("Add:", np.add(a, b))
print("Subtract:", np.subtract(b, a))
print("Multiply:", np.multiply(a, b))
print("Divide:", np.divide(b, a))
print("Power:", np.power(a, 2))
print("Mod:", np.mod(b, 3))

c = np.array([1.2, 2.7, 3.5])

print("Round:", np.round(c))
print("Ceil:", np.ceil(c))
print("Floor:", np.floor(c))
```

```python
# Quick Comparison
Function	               Purpose	           Example Output

arange()	            Create sequence	          [1 2 3]
add()	                 Addition	              [5 7 9]
subtract()	            Subtraction	              [-3 -3 -3]
multiply()	            Multiplication	          [4 10 18]
divide()	            Division	              [2.0 2.5]
power()	                Exponent	              [4 9 16]
mod()	                Remainder	              [1 2 0]
round()	                Normal rounding	          [2. 5. 6.]
ceil()	                Round upward	          [2. 3. 4.]
floor()	                Round downward	          [1. 2. 3.]
```
