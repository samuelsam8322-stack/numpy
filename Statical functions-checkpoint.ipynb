# 1. Mean (np.mean())

The mean is the average value of all numbers.

```python
import numpy as np

arr = np.array([10, 20, 30, 40])

print(np.mean(arr))
```

# 2. Median (np.median())

The median is the middle value after sorting the data.

```python
arr = np.array([1, 3, 5])

print(np.median(arr))
```

```python
# EXAMPLE 2
arr = np.array([1, 2, 3, 4])

print(np.median(arr))
```

```python
Explanation

Middle two values are 2 and 3

(2+3)/2=2.5
```

# 3. Average (np.average())

average() is similar to mean().

```python
arr = np.array([10, 20, 30])

print(np.average(arr))
```

```python
arr = np.array([10, 20, 30])

print(np.average(arr))
```

```python
# Difference Between Mean and Average

mean() → simple average
average() → can also use weights
```

# 4. Sum (np.sum())

Adds all elements.

```python
arr = np.array([1, 2, 3, 4])

print(np.sum(arr))
```

# 5. Percentile (np.percentile())

A percentile tells the value below which a certain percentage of data falls.

```python
arr = np.array([10, 20, 30, 40, 50])

print(np.percentile(arr, 25))
print(np.percentile(arr, 50))
print(np.percentile(arr, 75))
```

# 6. Unique (np.unique())

Returns only unique values.

```python
arr = np.array([1, 2, 2, 3, 4, 4, 5])

print(np.unique(arr))
```

# 7. Standard Deviation (np.std())

Standard deviation measures how spread out data is from the mean.

Small std → values are close to mean
Large std → values are spread out

```python
arr = np.array([10, 20, 30, 40])

print(np.std(arr))
```

# 8. Variance (np.var())

Variance is the square of standard deviation.

```python
arr = np.array([10, 20, 30, 40])

print(np.var(arr))
```

```python
# Difference Between Variance and Standard Deviation
Function	    Meaning
np.var()	Measures spread using squared units
np.std()	Square root of variance
```

```python
# All Functions Together Example
import numpy as np

arr = np.array([10, 20, 20, 30, 40])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Average:", np.average(arr))
print("Sum:", np.sum(arr))
print("Percentile 50:", np.percentile(arr, 50))
print("Unique:", np.unique(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
```

```python
Quick Summary Table
Function	           Purpose

np.mean()	         Average value
np.median()	         Middle value
np.average()	     Weighted/simple average
np.sum()	         Total sum
np.percentile()	     Percentage-based value
np.unique()	         Remove duplicates
np.std()	         Spread of data
np.var()	         Squared spread
```
