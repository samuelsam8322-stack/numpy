# ARRAY JOINING:

In NumPy, array joining means combining two or more arrays into one array.
Common methods are:

concatenate()
stack()
vstack()
hstack()
using transpose() with joining

# 1. concatenate()

Joins arrays along an existing axis.

```python
np.concatenate((arr1, arr2), axis=0)
```

```python
# Example 1: 1D Array
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.concatenate((a,b))

print(result)
```

```python
# Example 2: 2D Array Row-wise
a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

result = np.concatenate((a,b), axis=0)

print(result)
```

```python
# Example 3: 2D Array Column-wise
result = np.concatenate((a,b), axis=1)

print(result)
```

# 2.Using concatenate() with transpose()

Sometimes arrays cannot join because dimensions do not match.
We use transpose to fix shape problems.

```python
a = np.array([[1,2,3]])
b = np.array([[4,5,6]])

# transpose b
b = b.T

result = np.concatenate((a.T, b), axis=1)

print(result)
```

# 3.vstack() (Vertical Stack)

Stacks arrays vertically (top to bottom).

```python
concatenate(axis=0)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.vstack((a,b))

print(result)
```

# 4.hstack() (Horizontal Stack)

Stacks arrays horizontally (side by side).

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.hstack((a,b))

print(result)
```

```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

result = np.hstack((a,b))

print(result)
```

# Difference Between vstack() and hstack()

Function	   Meaning	          Direction
vstack()	Vertical stack	    Top → Bottom
hstack()	Horizontal stack	Left → Right

```python
Important Rule

For joining arrays:

 1.Rows must match for horizontal joining
 2.Columns must match for vertical joining
```

```python
# Quick Memory Trick
Function	Easy Meaning 

concatenate()	   General joining
vstack()	       Join vertically
hstack()	       Join horizontally
transpose()	       Swap rows & columns
```

# ARRAY SPLIT

In NumPy, splitting means dividing one array into multiple smaller arrays.

# 1. split()

Splits an array into equal parts.


```python
# Syntax
np.split(array, sections)
```

```python
import numpy as np

a = np.array([1,2,3,4,5,6])

result = np.split(a, 3)

print(result)
```

```python
# Error Example
a = np.array([1,2,3,4,5])

np.split(a, 3)
```

# 2. array_split()

Works like split() but allows unequal division.

```python
a = np.array([1,2,3,4,5])

result = np.array_split(a, 3)

print(result)
```

```python
#  Difference Between split() and array_split()
Function	          Equal Division Required?
split()	                       Yes
array_split()	                No
```

# 3. vsplit() (Vertical Split)

Splits arrays vertically (row-wise).

Works only for 2D or higher arrays.

```python
a = np.array([[1,2],
              [3,4],
              [5,6],
              [7,8]])

result = np.vsplit(a, 2)

print(result)
```

# 4. hsplit() (Horizontal Split)

Splits arrays horizontally (column-wise).

```python
a = np.array([[1,2,3,4],
              [5,6,7,8]])

result = np.hsplit(a, 2)

print(result)
```

```python
# Important Rules
Function	        Split Direction

split()	            General split
array_split()	    Unequal split allowed
vsplit()	        Row-wise split
hsplit()	        Column-wise split
```
