```python
In NumPy, the random module is used to generate random numbers, random arrays, and perform random operations.
It is very useful in data analytics, simulations, machine learning, and testing.
```

```python
np.random.rand(rowS, columns)
```

```python
import numpy as np

a = np.random.rand(2,3)
print(a)
```

```python
# 2. randn()

Generates random numbers from a normal distribution.

Mean = 0
Standard deviation = 1
```

```python
np.random.randn(rows, columns)
```

```python
a = np.random.randn(2,2)
print(a)
```

# 1. randint()

```python
Generates random integers between a range.
```

```python
np.random.randint(start, end, size)
```

```python
a = np.random.randint(1, 10, 5)
print(a)
```

# 2. random()

```python
Generates random float numbers between 0 and 1.
```

```python
np.random.random(size)
```

```python
a = np.random.random((2,2))
print(a)
```

# 3. seed()

```python
Used to generate the same random numbers every time.
```

```python
np.random.seed(1)

print(np.random.randint(1,10,5))
```

# 4. shuffle()

```python
Shuffles original array.
```

```python
a = np.array([1,2,3,4,5])

np.random.shuffle(a)

print(a)
```

# 5. choice()

```python
Selects random values from array.
```

```python
a = np.random.choice([10,20,30,40], 3)

print(a)
```

```python
# Simple Practice Program

import numpy as np

# Random integers
a = np.random.randint(1,100,10)

print("Random Numbers:")
print(a)

# Shuffle
np.random.shuffle(a)

print("After Shuffle:")
print(a)
```

```python
#Summary Table
  Function	                             Purpose
rand()	                      Random floats between 0 and 1
randn()	                      Random values from normal distribution
randint()	                  Random integers
random()	                  Random float array
seed()	                      Repeat same random values
shuffle()	                  Shuffle original array
permutation()	              Shuffled copy
uniform()	                  Random numbers in range
normal()	                  Normal distribution values
choice()	                  Random selection
```
