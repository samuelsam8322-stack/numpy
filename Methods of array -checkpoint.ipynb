# METHODS OF ARRAY IN NUMPY:

# 1. Copy in NumPy

A copy creates a completely separate array in memory.

Changes in the copied array do NOT affect the original array.

```python
import numpy as np

a = np.array([1, 2, 3])

b = a.copy()

b[0] = 100

print(a)   # [1 2 3]
print(b)   # [100 2 3]
```

```python
a  ----> [1 2 3]

b  ----> [1 2 3]   (separate memory)
```

# 2. View in NumPy

A view shares the same memory with the original array.

Changing the view also changes the original array.

```python
a = np.array([1, 2, 3])

b = a.view()

b[0] = 100

print(a)   # [100 2 3]
print(b)   # [100 2 3]
```

```python
a ----\
       ---> same memory
b ----/
```

# Difference Between Copy and View

Feature	                Copy	         View

Separate memory     	Yes              No
Changes affect original	No	             Yes
Faster	               Slower	         Faster
Memory usage	        More	         Less

# 3. Flatten

flatten() converts a multi-dimensional array into a 1D array.

It returns a copy.

```python
a = np.array([[1, 2],
              [3, 4]])

b = a.flatten()

print(b)
```

```python
Since flatten gives a copy: Original array does NOT change.
```

```python
b[0] = 100

print(a)
```

# 4. Ravel

ravel() also converts array into 1D.

But it returns a view whenever possible.

```python
a = np.array([[1, 2],
              [3, 4]])

b = a.ravel()

print(b)
```

```python
# Flatten vs Ravel
Feature	              flatten()	       ravel()

Returns	                Copy	       View (mostly)
Memory efficient	     No	           Yes
Faster	                Slower	       Faster
Original affected	    No	           Yes
```

# 5. Reshape

reshape() changes array dimensions without changing data.

Example 1: 1D → 2D

```python
a = np.array([1, 2, 3, 4, 5, 6])

b = a.reshape(2, 3)

print(b)
```

```python
a = np.array([1,2,3,4,5,6])

b = a.reshape(3, -1)

print(b)
```

```python
Total elements must remain same.
```

# 6. Transpose

Transpose converts rows into columns and columns into rows.

```python
transpose()
```

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a.T)
```

```python
Before
2 × 3
After transpose
3 × 2
```

# 7. Broadcasting

Broadcasting allows NumPy to perform operations on arrays of different shapes.

This is one of NumPy’s most powerful features.

```python
a = np.array([1, 2, 3])

print(a + 10)
```

```python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a + b)
```

```python
#  Broadcasting Rules

Two dimensions are compatible when:

They are equal
OR
One of them is 1
```
